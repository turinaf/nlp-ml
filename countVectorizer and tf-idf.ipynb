```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
def document_term_matrix(doc_list, vectorizer):
    doc_term_matrix = vectorizer.fit_transform(doc_list)
    return pd.DataFrame(doc_term_matrix.toarray(), columns=vectorizer.get_feature_names_out())
```

```python
# Considering a single senteces as one document.
doc_list = ["Just learning text preprocessing", 
            "For machine learning and Deep learning",
           "Cheers to more learning in the future."]
```

```python
count_vect = CountVectorizer()
document_term_matrix(doc_list, count_vect)
```

```python
tfidf_vect = TfidfVectorizer()
document_term_matrix(doc_list, tfidf_vect)
```
